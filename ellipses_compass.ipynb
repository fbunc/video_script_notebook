# Derivation of an expression for an Ellipse using complex numbers



$$\left(\frac{x}{x_o}\right)^2+\left(\frac{y}{y_o}\right)^2=1$$

$$x_t=x_o \cos{t}$$

$$y_t=y_o \sin{t}$$

$$z_t=x_t+iy_t$$

$$z_t=x_o \cos{t}+iy_o \sin{t}$$

$$z_t=x_o\frac{e^{it}+e^{-it}}{2}+iy_o \frac{e^{it}-e^{-it}}{2i}$$

$$z_t=x_o\frac{e^{it}+e^{-it}}{2}+iy_o \frac{e^{it}-e^{-it}}{2i}$$

$$z_t=x_o\frac{e^{it}+e^{-it}}{2}+y_o \frac{e^{it}-e^{-it}}{2}$$

$$z_t=x_o\frac{e^{it}}{2}+x_o\frac{e^{-it}}{2}+y_o \frac{e^{it}}{2}-y_o \frac{e^{-it}}{2}$$

$$z_t=\frac{x_o+y_o}{2} e^{it} + \frac{x_o-y_o}{2} e^{-it}$$

$$z_t= R_{_+} e^{it} +R_{_-} e^{-it}$$

Assuming $x_o \geq y_o$ the foci  $F_1$ and $F_2$ are:



$$F_{_1}=\sqrt{x_o^2-y_o^2}$$

$$F_{_2}=-F_{_1}$$

Or in terms of $R_{_+}$ and $R_{_-}$ :

$$F_{_1}=2\sqrt{R_{_+}R_{_-}}$$

Given that:

$$\sqrt{R_{_+}R_{_-}}=\sqrt{\frac{(x_o+y_o)}{2}\frac{(x_o-y_o)}{2}}$$

$$\sqrt{R_{_+}R_{_-}}=\sqrt{\frac{x_o^2-y_o^2}{4}}$$

$$2\sqrt{R_{_+}R_{_-}}=\sqrt{x_o^2-y_o^2}$$

## Oval expression used in Tau planet:

With $R_{_+},R_{_-} \in \mathbb{C}$

$$z_t= R_{_+} e^{it} +R_{_-} e^{-it}$$

$$F_{\pm}=\pm 2\sqrt{R_{_+}R_{_-}}$$



* We have an ellipse only when certain conditions are met. For example assuming $x_o>y_o we have

$$\left(\frac{x}{x_o}\right)^2+\left(\frac{y}{y_o}\right)^2=1$$

$$R_{_+}=\frac{x_o+y_o}{2}$$

$$R_{_-}=\frac{x_o-y_o}{2}$$


$$F_{\pm}=\pm 2\sqrt{R_{_+}R_{_-}}$$

* We have a circumference when $R_{_-}=0$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100)
x_o = 2
y_o = 1
sx = 14
sy = 10
R_p = (x_o + y_o) / 2
R_n = (x_o - y_o) / 2
P_t = R_p * np.exp(1j * t)
N_t = R_n * np.exp(-1j * t)
z_t = P_t + N_t

f1 = np.sqrt(x_o**2 - y_o**2)
f2 = -np.sqrt(x_o**2 - y_o**2)

dist_f1_zt = np.zeros_like(z_t, dtype='float')
dist_f2_zt = np.zeros_like(z_t, dtype='float')
conserved_dist = np.zeros_like(z_t, dtype='float')

for n in np.arange(z_t.size):
    dist_f1_zt[n] = np.abs(z_t[n] - f1)
    dist_f2_zt[n] = np.abs(z_t[n] - f2)
    conserved_dist[n] = dist_f1_zt[n] + dist_f2_zt[n]

# Plotting the arrays
plt.figure(figsize=(sx, sy))
plt.subplot(131)
plt.plot(dist_f1_zt, label='Distance $F_1$ to z_t')
plt.title('Distance from $F_1$ to z_t')
plt.legend()

plt.subplot(132)
plt.plot(dist_f2_zt, label='Distance $F_2$ to z_t')
plt.title('Distance from $F_2$ to z_t')
plt.legend()

plt.subplot(133)
plt.plot(conserved_dist, label='Conserved distance')
plt.title('Conserved distance')
plt.legend()

plt.tight_layout()
plt.show()
```
